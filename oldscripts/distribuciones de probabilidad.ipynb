# Ejemplo Diferentes Distribuciones de Probabilidad Discretas



```python
import numpy as np
import matplotlib.pyplot as plt
```

## Poisson



```python
lmbd = 0.5
n = 50000
s =np.random.poisson(lmbd,n)
```

```python
count, bins, ignored = plt.hist(s, max(s), density=False)
plt.show()
```

## Geometric (Recommended Distribution to test first) 

```python
z = np.random.geometric(p=0.05,size=n)
```

```python
count, bins, ignored = plt.hist(z, max(z), density=True)
```

## Multinomial

```python
z= np.random.multinomial(2000, [1/6.]*6, size=n)
```

```python
count, bins, ignored = plt.hist(z)
```
