```python
def PlotH(x,array):
    "Imprime un histograma dado dos arreglos"
    plt.figure(figsize = [10,10])
    plt.xlabel('Número de datos consecutivos faltantes')
    plt.ylabel('Cantidad de datos faltantes')
    plt.title('Histograma de los primeros {} datos faltantes'.format(len(x)))
    plt.bar(x,array)
    plt.show()

def PlotV(array):
    "Imprime puntos dado una arreglo"
    x = list(range(0,len(array)))
    plt.figure(figsize=[10,10])
    plt.xlabel('Tiempo (horas) 1986-2018')
    plt.ylabel("O3 medido")
    plt.title("Niveles de O3")
    #plt.stem(x,array)
    plt.plot(x,array,'bo')
    plt.show()
    
#Conteo de datos faltantes
def Nisnan(dataA):
    tmpN = 0
    for i in dataA:
        if mt.isnan(i):
            tmpN += 1
    return tmpN


#A la entrada un vector de indices a la salida el tamaño de los indices continuos
def Vecnt(dataA):
    tmpA = []
    lix = []
    for k,g in groupby(enumerate(dataA), lambda ix : ix[0] - ix[1]):     
        lix = list(map(itemgetter(1), g))
        print(lix)
        tmpA.append(len((list(map(itemgetter(1), g)))))
    return tmpA

#A la entrada un vector de datos a la salida la cantidad de nan
def Cnan(dataA):
    tmpV = []
    for j in range(len(dataA)):
        if(mt.isnan(dataA[j])):
            tmpV.append(j)
    return tmpV
```
